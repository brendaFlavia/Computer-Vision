```python
#Roberts Edge detector

roberts_cross_v = np.array( [[1, 0 ],
							[0,-1 ]] )

roberts_cross_h = np.array( [[ 0, 1 ],
							[ -1, 0 ]] )
path = r'C:\Users\DELL\Downloads\Assignment 2 Folder-20220719\Experiental images\frame18756.jpg'

imgr = cv2.imread(path,0).astype('float64')
imgr/=255.0
vertical = ndimage.convolve( imgr, roberts_cross_v )
horizontal = ndimage.convolve( imgr, roberts_cross_h )

edged_img = np.sqrt( np.square(horizontal) + np.square(vertical))
edged_img*=255
#cv2.imwrite("output.jpg",edged_img)

#Display Original Image
plt.figure(figsize=(18,19))
plt.subplot(121)
plt.imshow(imgr)
plt.title('my original Image')
plt.axis("off")


#Display Roberts Edge Detected Image
plt.subplot(122)
plt.imshow(edged_img)
plt.title('Roberts Edge Detected Image')
plt.axis("off")

plt.show()
```

```python
#Robert Edge Detector
#Advantages:
#1.Detection of edges and orientation are very easy
#2.Diagonal direction points are preserved

#Limitations:
#1.Very sensitive to noise
#2.Not very accurate in edge detection
```

```python
#Prewitt Edge Detector


# fetch image
prewit_path = r'C:\Users\DELL\Downloads\Assignment 2 Folder-20220719\Experiental images\frame18756.jpg'
img_prewit = cv2.imread(prewit_path, cv2.COLOR_BGR2GRAY)
rgb_img_prewit = cv2.cvtColor(img_prewit, cv2.COLOR_BGR2RGB)

# Grayscale image processing
grayImage_prewit = cv2.cvtColor(img_prewit, cv2.COLOR_BGR2GRAY)

# Prewitt operator
kernelx = np.array([[1, 1, 1], [0 ,0 ,0], [- 1, - 1,- 1]],dtype=int)
kernely = np.array([[- 1, 0, 1], [- 1 ,0 ,1], [- 1, 0, 1]],dtype=int)

x = cv2.filter2D(grayImage_prewit, cv2.CV_16S, kernelx)
y = cv2.filter2D(grayImage_prewit, cv2.CV_16S, kernely)

# turn to uint8, image fusion
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
Prewitt = cv2.addWeighted(absX, 0.5, absY, 0.5,0)



# Display graphics
titles = ['Original image','Prewitt operator']
images = [rgb_img_prewit, Prewitt]

for i in range(2):
    plt.subplot(1,2, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Display Original Image
plt.figure(figsize=(18,19))
plt.subplot(121)
plt.imshow(rgb_img_prewit)
plt.title('my original Image')
plt.axis("off")


#Display Roberts Edge Detected Image
plt.subplot(122)
plt.imshow(Prewitt)
plt.title('Prewitt operator')
plt.axis("off")

plt.show()
```

```python
#Advantages:
#1.Good performance on detecting vertical and horizontal edges
#2.Best operator to detect the orientation of an image

#Limitations:
#1.The magnitude of coefficient is fixed and cannot be changed
#2.Diagonal direction points are not preserved always
```
